## Read in Fip Code Data

```python
import pandas as pd
```

```python
fips_data_path = "../../data/county_level/county_fips.csv"
df = pd.read_csv(fips_data_path)
```

```python
df.head(5)
```

```python
df_copy = df.copy()
```

```python
df_copy["FIPS"] = df["FIPS"].apply(str)
df_copy.head()
```

```python
def get_last_three(num):
    return num[-3:]
```

```python
df_copy["FIPS"] = df_copy["FIPS"].apply(get_last_three)
df_copy.head()
```

## Get Data

```python
df_subset = df_copy[['State','FIPS']]
df_subset.head()
```

## Real Download

```python
import wget
```

```python
url_first_half = "https://www.ncdc.noaa.gov/cag/county/time-series/" # + AL-001
url_second_half = "-tavg-all-12-2010-2016.csv"
```

```python
def download(row):
    fips = row["FIPS"]
    state = row["State"]
    url = url_first_half + state + "-" + fips + url_second_half
    print("downloading: " + url)
    file_name = wget.download(url, out="../../data/county_level/county_climate_temp/")
```

```python
download(df_subset.loc[0])
```

```python
df_subset.apply(download, axis = 1)
```
