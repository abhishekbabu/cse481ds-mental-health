```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/processed_data/full_df.csv")
df.head()
```

```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
df.head()
```

```python
rucc_df = pd.read_excel('../../data/county_level/urbanization_codes/rucc2013.xls')
rucc_df.head()
```

```python
rucc_df['FIPS'] = rucc_df['FIPS'].astype('str').apply(lambda fips: "0" + fips if len(fips) == 4 else fips)
rucc_df.head()
```

```python
rucc_df.info(verbose=True)
```

```python
df['5-digit FIPS Code'] = df['5-digit FIPS Code'].astype('str').apply(lambda fips: "0" + fips if len(fips) == 4 else fips)
df.head()
```

```python
# Create dict of county FIPS codes to 2013 RUCC code for quick access
rucc_dict = dict(zip(rucc_df.FIPS, rucc_df.RUCC_2013))
```

```python
def compute_rucc_code(row):
    rucc_2013 = rucc_dict[row['5-digit FIPS Code']]
    return rucc_2013
    """
    pop = row['Population']
    if row['Release Year'] == 2013:
        return rucc_2013
    elif rucc_2013 <= 3: # Metropolitan counties
        if pop >= 1000000:
            return 1.0
        elif pop >= 250000:
            return 2.0
        else:
            return 3.0
    elif rucc_2013 in {4.0, 6.0, 8.0}: # Nonmetroplitan counties adjacent to metro areas
        if pop >= 20000:
            return 4.0
        elif pop >= 2500:
            return 6.0
        else:
            return 8.0
    else: # Nonmetropolitan counties not adjacent to metro areas
        if pop >= 20000:
            return 5.0
        elif pop >= 2500:
            return 7.0
        else:
            return 9.0
    """
```

```python
df['RUCC'] = df.apply(compute_rucc_code, axis=1)
df.head()
```

```python
df.groupby(['5-digit FIPS Code', 'State Abbreviation', 'Name']).Population.apply(list).reset_index() # list of population values for each county
```

```python
df.groupby(['5-digit FIPS Code', 'State Abbreviation', 'Name']).RUCC.apply(list).reset_index() # list of corresponding computed RUCC values for each county
```

```python
df.groupby(['5-digit FIPS Code', 'State Abbreviation', 'Name'])['Release Year'].apply(list).reset_index() # list of corresponding computed RUCC values for each county
```
