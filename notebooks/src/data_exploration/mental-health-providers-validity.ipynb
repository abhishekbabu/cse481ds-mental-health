### 2. Convergence Validity
Code source: https://stackoverflow.com/questions/25579227/seaborn-implot-with-equation-and-r2-text

```python
def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data["Ratio of mental health providers to population"], data['Poor mental health days raw value'])
    ax = plt.gca()
    ax.text(.05, .8, 'r={:.2f}, p={:.2g}'.format(r, p),
            transform=ax.transAxes)
```

```python
def invert_pop_mental_health_ratio(row):
    ratio = row["Ratio of population to mental health providers"]
    if ratio < 0:
        return 0
    return 1 / ratio
```

```python
df_plot = df_mental_provider[["Ratio of population to mental health providers", 'Poor mental health days raw value', 'Release Year']].replace([np.inf, -np.inf], np.nan).dropna()
df_plot["Ratio of mental health providers to population"] = df_plot.apply(invert_pop_mental_health_ratio, axis=1)
g = sns.lmplot(x="Ratio of mental health providers to population", y="Poor mental health days raw value", col="Release Year", data=df_plot)
g.map_dataframe(annotate)
plt.show()
```

```python
years = sorted(df_mental_provider['Release Year'].unique()) # get unique year
colors = plt.get_cmap('tab10') # Tableau 10 color scheme
fig, axes = plt.subplots(2, (len(years) + 1) // 2, figsize=(15, 7.5), sharey=True) # create subplot grid
for j, ax in enumerate(axes.flatten()):
    if j >= len(years):
        fig.delaxes(ax) # remove all extra axes
    else:
        sns.boxplot(x="RUCC", y="Mental health providers raw value", ax=ax, data=df_mental_provider[df_mental_provider['Release Year'] == years[j]])
        ax.set_title(years[j])
fig.suptitle('Yearly Box plot by RUCC Code')
plt.show()
```
